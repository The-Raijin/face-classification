## Load image and do basic colour space manipulation

```python
# Read image from relative path
image = cv2.imread('../Images/Image-001.jpg')

# Plot read image
plt.figure(figsize=(20,10))
plt.imshow(image)

# Show image size on the top of the figure
print("height, weight, channels = ", image.shape)
```

- Standard image colour space is BGR, so it creates and error when shown by matplotlib, which shows it in a RGB colour space, therefore creating the effect above. <br>
- To correct this we have to convert the image to the RGB colour space. <br>
- Then we use the following code:

```python
# Convert BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Plot image
plt.figure(figsize=(20,10))
plt.imshow(image_rgb)

# Show image size on the top of the figure
print("height, weight, channels = ", image_rgb.shape)
```

- We can also convert the image to gray scale by using similar code:

```python
# Convert RGB to Gray scale
image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)

# PLot image
plt.figure(figsize=(20,10))
plt.imshow(image_gray, cmap='gray')

# Show image size on the top of the figure
print("height, weight, channels = ", image_gray.shape)
```

## Define region of interest (ROI) using Slicing Notation

We can clip one part of the image by selecting rows and columns of the image:

```python
# Select ROI isong Slicing Notation
image_roi = image_rgb[1300:2650,2200:2800,:] 

# Plot an specific image ROI
plt.figure(figsize=(20,10))
plt.imshow(image_roi)

# Show image size on the top of the figure
print("height, weight, channels = ", image_roi.shape)
```

## Applying learned functions to an image (playing with ROI and colour spaces)

- And we can combine different colour spaces in new images to generate cool effects. On this example, I want my ROI to be RGB and the rest of the image will be gray scale.

```python
# Create 3 channel image
final_image = image_rgb.copy()

# Fill each channel with a gray scale image
final_image[:,:,0] = image_gray.copy()
final_image[:,:,1] = image_gray.copy()
final_image[:,:,2] = image_gray.copy()

# Fit colour image in the desired ROI
final_image[1300:2650,2200:2800,:] = image_rgb[1300:2650,2200:2800,:]
```

```python
# Show final image
plt.figure(figsize=(20,10))
plt.imshow(final_image)

# Show image size on the top of the figure
print("height, weight, channels = ", final_image.shape)
```

## Save image in a folder

```python
# Convert image to BGR standard
final_image_bgr = cv2.cvtColor(final_image, cv2.COLOR_BGR2RGB)

# Save it to
cv2.imwrite('../Images/Assignment Images/lecture-001-final-image.png', final_image_bgr)
```
